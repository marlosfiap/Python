```python
# Os operadores in e not in em Python são usados 
# para verificar se um valor está (ou não está)
# presente dentro de uma sequência ou coleção.
```

```python
#🔹 in (está contido?)
# Retorna True se o valor estiver dentro da estrutura.

lista = [1, 2, 3]

print(2 in lista)  # True
print(5 in lista)  # False
```

```python
# 🔹 not in (não está contido?)
# É o oposto: retorna True se o valor não estiver presente.7

lista = [1, 2, 3]

print(5 not in lista)  # True
print(2 not in lista)  # False
```

# 🧠 Onde funciona?
# Esses operadores funcionam com vários tipos de dados:

Listas (list)
Tuplas (tuple)
Strings (str)
Dicionários (dict)
Conjuntos (set)

```python
texto = "Python"

print("Py" in texto)   # True
print("Java" in texto) # False
```

```python
dados = {"nome": "Ana", "idade": 25}

print("nome" in dados)      # True (verifica a chave)
print("Ana" in dados)       # False (não verifica valores)
```

```python
print("Ana" in dados.values())  # True
```
