```python
# Os operadores de identidade em Python — is e is not — serve
# para comparar se duas variáveis apontam para o mesmo objeto
# na memória, e não apenas se têm o mesmo valor.
```

```python
curso = "Data Science"
nome_curso = curso
saldo, limite = 200, 200
```

```python
#Objeto A igual a objeto B
curso is nome_curso 
```

```python
# É o oposto: retorna True se são objetos diferentes.
curso is not nome_curso
```

```python
saldo is limite
```

```python
# ⚠️ Diferença importante: is vs ==
# == compara valores
# is compara identidade (memória)
```
